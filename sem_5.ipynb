**4. Есть ли статистически значимые различия в росте дочерей? Рост матерей 172, 177, 158, 170, 178,175, 164, 160, 169, 165 Рост взрослых дочерей: 173, 175, 162, 174, 175, 168, 155, 170, 160, 163**

```python
import numpy as np
from scipy import stats
x1=np.array([172, 177, 158, 170, 178, 175, 164, 160, 169, 165])
x2=np.array([173, 175, 162, 174, 175, 168, 155, 170, 160, 163])
```

```python
stats.ttest_rel(x1,x2)
```

Статистически значимых различий в росте матерей и дочерей нет
